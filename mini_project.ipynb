```python
import pandas as pd #open-source python library
print("Libraries imported successfully")
```

```python
df=pd.read_csv('student_performance.csv')
```

```python
print("DATASET OVERVIEW")
print("\nTotal students")
print(df.shape[0])
print("\nTotal Columns")
print(df.shape[1])
length=df['department'].unique()
print(f'Departments {len(length)}')

print("\n")
print("[2] DEPARTMENT-WISE STUDENT COUNT")
print(df['department'].value_counts())

print("\n")
print("Highest Math Score: ",df.math_score.max())
print("Lowest Math Score: ",df.math_score.min())
print("Average Math Score: ", df.math_score.mean().round(2))
print(f"Average Programming Score: {df.programming_score.mean():.2f}")
print("Average Attendance: ", df.attendance_percentage.mean().round(2), "%")

df['total_score']=df['math_score']+df['programming_score']+df['science_score']+df['english_score']
g=df.groupby('gender')['total_score'].mean().round(2)
print(g)


top5=df.sort_values('total_score',ascending=False).head(5)
for index, row in top5.iterrows():
  print(f"{row['name']:<20} | {row['department']:<18} | Score:{row['total_score']}")

print("\n")
low_att=df[df['attendance_percentage']<75][['name','department','attendance_percentage']]
if len(low_att)>0:
  print(low_att.to_string(index=False))
else:
  print("All students have attendance above 75%")

print("Report complete")
```
